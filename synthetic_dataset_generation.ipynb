# Loading labeled data and general dataset

```python
labeled_court_cases = pd.read_parquet('data/court_cases_labeled.parquet')
labeled_court_cases.head(2)
```

```python
all_court_cases = pd.read_parquet('data/2024-supreme-court-decisions-deduplicated.parquet')
all_court_cases.head(2)
```

```python
# ensure all columns are strings
labeled_court_cases = labeled_court_cases.astype(str)
all_court_cases = all_court_cases.astype(str)
# merge the two datasets
merged_data = pd.merge(labeled_court_cases, all_court_cases, on='id', how='inner')

print("There are " + str(len(merged_data)) + " documents that are present in both the labeled and unlabeled datasets.")
print("Removing documents from unlabeled dataset that are present in the labeled dataset to avoid data leakage...")

all_court_cases = all_court_cases[~all_court_cases['id'].isin(labeled_court_cases['id'])]

print("Done! Verifying once again...")

merged_data = pd.merge(labeled_court_cases, all_court_cases, on='id', how='inner')
print(f"There are {len(merged_data)} documents that are present in both the labeled and unlabeled datasets.")
print(f"Length of unlabeled dataset: {len(all_court_cases)}")
print(f"Length of labeled dataset: {len(labeled_court_cases)}")
```
